```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df=df.dropna()
```

```python
#to binarize the relative humidity to 0 or 1
```

```python
#a copy of the clean data is created to save operation
clean_data=df.copy()
```

```python
clean_data
```

```python
clean_data['high_humidity_label']=(clean_data['relative_humidity_3pm']>24.99)*1
print(clean_data['high_humidity_label'])
#the value 24.99 which has been used to convert, it will be given in question
```

```python
clean_data
```

```python
y=clean_data['high_humidity_label']
```

```python
y
```

```python
morning_features=clean_data.drop(['high_humidity_label','relative_humidity_3pm','relative_humidity_9am','number'],axis=1)
```

```python
df.columns
```

```python
X=morning_features
```

```python
X1=morning_features.columns
```

```python
X1
```

```python
X
```

```python
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=3)
```

```python
X_train.head()
```

```python
y_test
```

```python
X_test
```

```python
humidity_classifier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
humidity_classifier.fit(X_train,y_train)
```

```python
type(humidity_classifier)
```

```python
predictions=humidity_classifier.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(100,100))
_= tree.plot_tree(humidity_classifier,
                 feature_names=X1,
                 class_names="high_humidity_label",
                 filled=True)
```
